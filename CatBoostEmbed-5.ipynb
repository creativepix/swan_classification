```python
def pickle_to_df(vc):
  cls,embeds = [],[]
  for key in vc.keys():
    for emb in vc[key]:
      cls += [emb[0]]
      embeds += [emb[1]]
  df = pd.DataFrame()
  df['cls'] = cls
  df['embeds'] = embeds
  return df
```

```python
def create_df(cl,embeds):
  df['cls'] = cl
  for i in range(embeds.shape[-1]):
    df[f'embed_{i}'] = embeds.T[i]
  return df
```
